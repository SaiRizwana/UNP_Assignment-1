```python
import numpy as np
import pandas as pd
```

Generate 10000 values between 5 and 500. Find its mean,median,variance,25th percentile.

```python
a=np.random.randint(5,500,10000)
print(a)
```

```python
print(np.mean(a))
print(np.var(a))
print(np.median(a))
print(np.percentile(a,25))
```

```python
data=pd.read_csv("dat.csv",index_col=0)
data
```

In the "dat.csv" replace all the null values with the mean of their respective columns(except Gender).



```python
data["class"].fillna(np.mean(data["class"]),inplace=True)
data["Age"].fillna(np.mean(data["Age"]),inplace=True)
data["Survived"].fillna(np.mean(data["Survived"]),inplace=True)
```

```python
data
```

Create a new column called "Bool" which has value as "Yes" if Survived=0 and "no" if Survived=1

```python
data['Bool'] = np.where(data['Survived']== 1, "Yes", "No")
data
```
